```python
from imagededup.methods import DHash
from glob import glob
from imagededup.utils import plot_duplicates
import random
import os
import shutil
```

```python
diff_hash = DHash()
```

```python
medias = "/home/bmazoyer/Documents/Data/police/medias/"
```

```python
output = "/home/bmazoyer/Documents/Data/police/medias_unique/"
```

```python
# First run on images from all subdirs.
all_duplicates = {}
for folder in glob(medias + "/*/"):
    if len(os.listdir(folder)) > 0:
        files_to_remove = set()
        duplicates = diff_hash.find_duplicates(image_dir=folder)
        for k, v in duplicates.items():
            if k not in files_to_remove:
                shutil.copyfile(folder + k, output + k)  
                tmp = [i[0] if isinstance(i, tuple) else i for i in v]  # handle tuples (image_id, score)
                files_to_remove.update(tmp)
                all_duplicates[k] = tmp
```

```python
len(all_duplicates)
```

```python
# Second run only on selected images
duplicates = diff_hash.find_duplicates(image_dir=output)
files_to_remove = {}
for k, v in duplicates.items():
    if k not in files_to_remove:
        for i in v:
            removed_file = i[0] if isinstance(i, tuple) else i
            files_to_remove[removed_file] = k
        
for file in files_to_remove:
    try:
        os.remove(output + file)
        all_duplicates[files_to_remove[file]].extend(all_duplicates[file])
        all_duplicates.pop(file)
    except Exception as e:
        print(e)
        continue
```
